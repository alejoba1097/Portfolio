# Linear Regressor

Code used to estimate the regression coefficients using Least Squares and Neural Networks.

```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
# Generate random data for regression
X = np.random.multivariate_normal(np.zeros(5),np.eye(5),1000)
X = np.concatenate((X, np.ones((1000,1))), axis = 1)

# B matrix with coefficients
B = np.random.uniform(-10,10,6)

# Dependent variable Y with random noise
Y = X@B + np.random.uniform(0, 1, 1000)
```

```python
input_size = 6
output_size = 1
model = tf.keras.Sequential([
    tf.keras.layers.Dense(output_size)
])
```

```python
custom_optimizer = tf.keras.optimizers.SGD(learning_rate=0.02)
model.compile(optimizer=custom_optimizer, loss = 'mean_squared_error')
model.fit(X, Y, epochs = 100, verbose = 1)
```

```python
# Compute loss
def squared_loss(Y, Y_hat):
    return sum((model(X) - Y.reshape((1000,1)))**2)
```

```python
# Define the model
inputs = tf.keras.Input(shape=(6,))
outputs =  tf.keras.layers.Dense(1, use_bias = False)(inputs)
model = tf.keras.Model(inputs = inputs, outputs = outputs)

learning_rate = 0.0001
for i in np.arange(1,1000):

    with tf.GradientTape() as tape:
        loss = squared_loss(Y, model(X))

    grad = tape.gradient(loss, model.trainable_variables)
    model.layers[1].set_weights(model.layers[1].get_weights() - grad[0]*learning_rate)
    print(loss)
    if loss < 0.01:
        break
```
